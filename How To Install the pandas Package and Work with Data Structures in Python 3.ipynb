```python
# Importing Libraries

import pandas as pd
import numpy as np
```

```python
# Creating a Series (1D data with labels)

## Example 1: A simple Series with automatic index

squares = pd.Series([0, 1, 4, 9, 16, 25], name="Squares")
print(squares)

## Example 2: A Series with custom labels

avg_ocean_depth = pd.Series(
    [1205, 3646, 3741, 4080, 3270],
    index=["Arctic", "Atlantic", "Indian", "Pacific", "Southern"]
)
print(avg_ocean_depth)

## Accessing values

print(avg_ocean_depth[2])          # Access by index number
print(avg_ocean_depth["Indian"])   # Access by label
```

```python
# Creating a DataFrame (2D data)

## Create two Series first

avg_depth = pd.Series({
    "Arctic": 1205,
    "Atlantic": 3646,
    "Indian": 3741,
    "Pacific": 4080,
    "Southern": 3270
})

max_depth = pd.Series({
    "Arctic": 5527,
    "Atlantic": 8486,
    "Indian": 7906,
    "Pacific": 10911,
    "Southern": 7075
})

## Combine into a DataFrame

ocean_depths = pd.DataFrame({
    "Avg. Depth (m)": avg_depth,
    "Max. Depth (m)": max_depth
})

print(ocean_depths)
```

```python
# Working with DataFrames

## Sort data by column

print(ocean_depths.sort_values("Avg. Depth (m)"))

## Get basic statistics (mean, min, max, etc.)

print(ocean_depths.describe())

## Check for missing data

print(ocean_depths.isna())

## Fill missing values with 0

print(ocean_depths.fillna(0))

## Drop rows with missing values

print(ocean_depths.dropna())
```
